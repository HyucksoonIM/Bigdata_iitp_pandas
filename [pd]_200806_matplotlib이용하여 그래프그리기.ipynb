```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1. histogram

```python
N = 100000
bins = 30

x = np.random.randn(N) 
#plt.hist(x, bins, weights, density, cumulative, label) 
# x(필수), bins(막대 영역(bin)을 얼마나 채우는지 결정)
plt.hist(x, bins=bins)
plt.show()
```

```python
#subplots = 여러 그래프 한번에 보여주기 위함. (subplot도 동일하지만 문법에 있어서 subplots가 더 편리)
#plt.subplots(number of rows, number of cols, tight_layout=True: 그래프 위치와 라벨 위치를 좀 정리해줌)
#fig(전체 subplot), axs(전체 중 낱낱개)
fig, axs = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(3, 9)

#density=True: 가중치 데이터 정규화
#cumulative=True: bool 자료형을 입력받아 가중치 데이터 그래프 그림. 즉, 최댓값은 1
axs[0].hist(x, bins=bins, density=True, cumulative=True)
axs[1].hist(x, bins=bins, density=True)
plt.show()
```

### 2. pichart

explode: 파이에서 툭 튀어져 나온 비율\
qutopct: 퍼센트 표기\
shadow: 그림자 표기\
startangle: 파이 그래프 시작 위치

```python
labels = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = [0.3, 0, 0, 0, 0, 0]

patches, texts, autotexts = plt.pie(sizes,
                                   explode=explode,
                                   labels=labels,
                                   autopct='%1.1f%%', #소수점 아래 1자리까지 표시
                                   shadow=True,
                                   startangle=90)
plt.title('Samsung Pie', fontsize=15)

#스타일 지정하기
#case1. label 텍스트 스타일 적용
for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')
#case2. pie 위 텍스트에 대한 텍스트 스타일 적용
for t in autotexts:
    t.set_color('white')
    t.set_fontsize(18)
    
plt.show()
```

### 3. BOX PLOT

```python
#샘플 데이터 생성
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

#### 3-1. 기본 box plot 설정

```python
plt.boxplot(data)
plt.tight_layout()
plt.show()
```

<박스플롯 설정>
상한값: 75% 지점에서 1.5배\
75%\
50%\
25%\
하한값\
주황선: 평균값\
수염 밖의 값들: 노이즈 값들

```python
#ver=False: 축을 바꿀 수 있음
plt.title("Horizontal Box Plot", fontsize=15)
plt.boxplot(data, vert=False)
plt.show()
```

#### 3-2. outlier 마커 심볼과 컬러 변경

```python
outlier_marker = dict(markerfacecolor='r', marker='D')

plt.title('Changed Outlier Symbols', fontsize = 15)
plt.boxplot(data, flierprops=outlier_marker)
plt.show()
```
