```python
import pandas as pd
import numpy as np
df = pd.read_excel('C:/Users/admin/Desktop/Big_data/Pandas/korean-idol.xlsx')
df
```

<h2> 1.data type 변환하기

```python
#'키'의 데이터타입을 int로변환
df['키'].astype(int)
```

<h3> NaN값이 있으면 데이터 타입을 변경할 수 없다. fillna를 사용하여 NaN값을 채워주도록 하자.

```python
df['키'] = df['키'].fillna(-1)
```

```python
#'키'의 데이터타입을 int로변환(다시시도)
df['키'].astype(int)
```

<h3> datetime으로 바꾸기

```python
df['생년월일'] = pd.to_datetime(df['생년월일'])
df['생년월일']
```

```python
df.head()
```

```python
#dayofweek: 월요일 = 0
df['생년월일'].dt.dayofweek
df['생년월일'].dt.month
```

```python
df['생일_년'] = df['생년월일'].dt.year
df['생일_월'] = df['생년월일'].dt.month
df['생일_일'] = df['생년월일'].dt.day
df.head()
```

<h2> 2. value 수정하기

<h3> 1) 조건에 맞게 value 바꾸기

```python
df.loc[df['성별'] == '남자', '성별'] = 1
df.loc[df['성별'] == '여자', '성별'] = 0
df
```

```python
df
```

<h3> 2) 함수 사용하기

```python
def male_or_female(x):
    if x == 1:
        return 1
    elif x == 0:
        return 0
df['성별_NEW'] = df['성별'].apply(male_or_female) #함수를 쓸 때는 .apply(함수명)를 사용한다.
df
```

```python
def cm_to_brand(df):
    value = df['브랜드평판지수']/df['키']
    return value

df.apply(cm_to_brand, axis=1)
```

<h3> 3) lambda 함수 사용하기

```python
#1. 람다 함수로 만들어서 넣기
f = lambda x:1 if x == '남자' else 0
df['성별'].apply(f)
```

```python
#2. 람다 바로 넣기
#case1
df['키/2'] = df['키'].apply(lambda x: x/2)
#case2
df['키**2'] = df['키'].apply(lambda x: x**2)

df.head()
```
