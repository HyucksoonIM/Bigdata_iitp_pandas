```python
import pandas as pd
import numpy as np
df = pd.read_excel('C:/Users/admin/Desktop/Big_data/Pandas/korean-idol.xlsx')
df1 = pd.read_csv('C:/Users/admin/Desktop/Big_data/Pandas/korean-idol-2.csv')
```

<h2> concat: DataFrame 합치기

```python
df_copy = df.copy()
df1_copy = df1.copy()
```

<h3> 1.두 개의 dataframe 합치기 (append와 동일)

```python
# data 형태를 리스트로 합쳐준다. ([df, df_copy], sort=False: 순서 유지 / sort=True: column명이 오름차순 정렬)
df_concat=pd.concat([df, df_copy], sort=False)
#인덱스 리셋
df_concat.reset_index(drop=True) #drop=True: 기존의 index 삭제
```
