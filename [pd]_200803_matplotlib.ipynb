<h3> 단일 그래프 그리기

```python
#-4.5: 시작점 / 5: 끝점 / 0.5: step
x = np.arange(-4.5, 5, 0.5)
y=2*x**2
(x,y)
```

```python
plt.plot(x,y)
```

<h3> 다중 그래프 그리기


```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x+30
y3 = 4*x**2 +10
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
plt.plot(x,y1)
plt.figure()
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
x=np.arange(-10, 10, 1)
y1=x**2-2
y2=20*np.cos(x)**2
x
```

```python
plt.figure(1) #1번 그래프 지목
plt.plot(x,y1) #지정된 창에 그래프 생성
```

```python
plt.figure(2) #2번 그래프 지목
plt.plot(x,y2) #지정된 창에 그래프 생성
```

```python
plt.figure(1) #1번 그래프 지목
plt.plot(x,y1) #지정된 창에 그래프 생성
plt.figure(1) #2번 그래프 지목
plt.plot(x,y2) #지정된 창에 그래프 그림(그래프 생성_덮어쓰기)
```

```python
plt.figure(2) #2번 그래프 지목
plt.clf() #그래프 제거
plt.plot(x,y1) #2번 그래프에 그래프 그림
```
