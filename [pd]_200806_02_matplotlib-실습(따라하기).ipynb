# 1. Scatterplot

```python
np.random.rand(50)
```

```python
np.arange(50)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 1000
scatter = plt.scatter(x, y, c=colors, s=area)
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.scatter(x, y, s = area, cmap = 'blue', alpha=0.1)
plt.title('alpha = 0.1')

plt.subplot(132)
plt.scatter(x, y, s = area, cmap = 'red', alpha=0.5)
plt.title('alpha = 0.5')

plt.subplot(133)
plt.scatter(x, y, s = area, cmap = 'green', alpha=1)
plt.title('alpha = 1')

plt.show()
```

# 2. Barplot, Barhplot


```python
#plt.bar(x, height, width, align, alpha)
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [90, 60, 80, 50, 70, 40]

#figure size
plt.figure(figsize = (10, 4))
#수직 barplot
plt.bar(x, y, alpha=0.7, color='red')
plt.title('Subjects')
plt.ylabel('Grades')
plt.show()
```

```python
plt.figure(figsize = (7,4))
plt.barh(x, y, alpha=0.7, color='green')
plt.title('[barh]subjects')
plt.show()
```
