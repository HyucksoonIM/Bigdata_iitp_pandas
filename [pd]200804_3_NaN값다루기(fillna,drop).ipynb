<h2> 1. fillna(): NaN 값을 채워주는 함수

```python
#'키' column의 NaN 값 확인
df['키']
```

```python
#NaN 값에 -1 대입
#inplace=True : 변경 후, 저장하는 명령어

df['키'] = df['키'].fillna(0) # =df['키'].fillna(0, inplace=True)
```

```python
df['키']
```

```python
#키 값이 변경된 것을 확인
df
```

<h2> 2. row, column 지우기

<h3> 1) dropna(): NaN 값이 있는 row를 삭제

```python
#index 2 번이 삭제되는 것을 볼 수 있음.
#(axis=0) 또는 (): NaN이 포함된 row 삭제
#(axis=1): NaN이 포함된 column 삭제
#how='any': 하나라도 NaN이 있으면 드랍
#how='all': 모두 NaN이면 드랍
df.dropna(axis=0, how='any')
```

```python
df.dropna(axis=1, how='all')
```

```python
#row=10에 NaN 넣기
df.iloc[10] = np.NaN
df
```

```python
df.dropna(axis=0, how='all')
```

<h3> 2) drop_ duplicates(): 중복 데이터 제거

```python
#중복 데이터 제거 drop.duplicates(keep='first'/'last')
#first(default): 처음것을 살림 / last: 마지막 것을 살림
df['키'].drop_duplicates(keep='last')
```

```python
df.drop_duplicates('그룹', keep='last')
```

<h3> 3) drop(): row/column 지정하여 제거

```python
#column 제거하기
df.drop(['그룹','소속사'], axis=1)
```

```python
#index 번호로 row 제거
df.drop([3,5], axis=0)
```
