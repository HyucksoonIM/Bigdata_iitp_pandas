<h2> 4. 데이터 피봇 다루기

```python
#pivot은 index, columns, values를 갖는다. 
#values는 int, float 데이터 타입만 가능하며, values를 입력하지 않으면 int, float형의 모든 column이 표시된다.
#aggfunc = sum/mean/median 값 등을 계산해서 넣을 수 있다.
pd.pivot_table(df, index='소속사', columns='혈액형', values='키', aggfunc = np.mean)
```

<h4> Groupby와 함께 쓸 수 있는 메소드: count / sum / mean / var / std / min/max

```python
df.groupby('소속사').mean()
```

```python
df.groupby('혈액형')['키'].mean()
```

```python
df2 = df.groupby(['혈액형', '성별']).mean()
df2
```

```python
df2.unstack('혈액형')
```

```python
df2.unstack('성별')
```
